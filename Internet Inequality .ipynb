## Income Inequality

This analysis looks at the inequality in broadband on the basis of income. The data is from ONS. 

https://www.ons.gov.uk/peoplepopulationandcommunity/personalandhouseholdfinances/incomeandwealth/bulletins/householddisposableincomeandinequality/yearending2018

https://www.ons.gov.uk/peoplepopulationandcommunity/personalandhouseholdfinances/incomeandwealth/adhocs/10666internetaccessbandedbyincome

The data was processed into csv format. These csv files can be found in the repo. 

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Data set published Feb 2019. Data from UK Financial year 2018. 
incomes = pd.read_csv("income_data/uk_household_income_2018.csv")
```

# The UK Household income Distribution 


```python
income_bracket = incomes['Income bracket']
normalised_count = incomes['count']/sum(incomes['count'])


fig = plt.figure()
fig.set_size_inches(12,8)
width = 1 * (income_bracket[1] - income_bracket[0])

center = (income_bracket[:-1] + income_bracket[1:]) / 2
plt.bar(center, normalised_count, align='center', width=width, color = 'y', edgecolor = 'k')

x_ticks = ["£"+str(i) + "k" for i in np.arange(0,80,10)]

plt.xticks(np.arange(0,80,10), x_ticks);
plt.xlabel("Household Income")
plt.ylabel("Normalised Count")
```

# Income of those with Internet Access 

The plot below shows the income distribtion of those with internet access. The income distibution from above is also shown for comparison. 

Rebinning involved finding the percentage of population that lie in a given bin. This is done by integrating between the bin range, i.e. taking a sum of the normalised bin counts for the income range we're interested in. 

```python
broadband = pd.read_csv("broadband_access_by_income.csv")
```

```python
income_bracket_floats = []
income_brackets_revise = []
for income_bracket_str in broadband['income_bracket']:
    
    income_bracket_split = income_bracket_str.split("-")
    income_bin_labels = []
    
    for i in income_bracket_split:
        i_ = i.replace('£','')
        income_bin_labels.append(i_[:-3] +"." + i_[-4:-3] + "k")
        
        income_bracket_floats.append(i_)
    income_brackets_revise.append(income_bin_labels)

# Add a space for the 96k tick
income_brackets_revise[-1].append(" ")

# Add dashes between lower and higher end 
income_brackets_revise = [i[0] + " - "+i[1] for i in income_brackets_revise]

income_binning = np.array([int(income_bracket_floats[i]) for i in np.arange(0,16,2)])/1000

# Rebin by taking average
bins = []
rebinned_income_dist = []
for i,j in enumerate(income_binning):
    if i <7:
        # List of bin counts which will be merged
        grouped_income_counts = normalised_count[int(income_binning[i]):int(income_binning[i+1])]
        # Integrate bin values to find percentage of population within given income bracket. 
        rebinned_income_dist.append(sum(grouped_income_counts))

        
# Normalise
# income_dist = np.array(rebinned_income_dist)/sum(rebinned_income_dist)
internet_income_dist = broadband['individual_count']/sum(broadband['individual_count'])

rebinned_income_dist = np.array(rebinned_income_dist)

fig = plt.figure()
fig.set_size_inches(12,8)
width = 0.3
center = np.array(range(8))
plt.bar(center[:-1],rebinned_income_dist , align='center', width=width, color = 'b', edgecolor = 'k', label = "Income Distribution")

plt.bar(center + 0.5, internet_income_dist, align='center', width=width, color = 'y', edgecolor = 'k',alpha = 0.7, label = "Income Distribution of those with Internet")



plt.xticks(center + 0.25,income_brackets_revise)
plt.legend()
plt.xlabel("Income in Pounds")
plt.title("Internet Access & Income Distribution")
```

## Caveats
* It is crucial to remember that the plot above is a bar plot with unequal bin widths. Plotting like this skews the income distribution to the right. 
* This was done because the internet access data is available only through this unequal binning. 
* This **not** a probability distribution. This could equally be a pie chart.
